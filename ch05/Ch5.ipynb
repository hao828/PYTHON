### 作業 文字接龍

```python
txt = input("請輸入一個字串：")
strList = []
strList.append(txt)
msg = txt
count = 0
while count < 5:
    print("上一個字串是:"+msg)
    tmpTxt = input("請輸入-"+txt[-1]+"-開始的字串")
    if(tmpTxt[0]==txt[-1]):
        txt = tmpTxt
        strList.append(tmpTxt)
        msg = "-".join(strList)
    else :
        count +=1
    
```

### isinstance

```python
list4 = ['x',1,'y',2,'z',3,5.3,[8,9]]
print(list4)
sum = 0
for v in list4:
    if isinstance(v,int):
        sum += v
    elif isinstance(v,str):
        print("str:",v)
print(sum)
```

```python
list5 = ['ken',1,'joy',2,'linda',3]
list6 = [v for v in list5]
print(list6)
listNo = [v for v in list5 if isinstance(v,int)]
print(listNo)
listName = [v for v in list5 if isinstance(v,str)]
print(listName)
```

### Key 與 Value


```python
d2 = {"a":100,"b":200,"c":300}
for k in d2:
    print(f"{k}--{d2[k]}")
print()
for k2 in d2.keys():
    print(f"{k2}--{d2[k2]}")
print()
for v in d2.values():
    print(f"v:{v}")
```

```python
d3 = {"a":100,"b":200,"c":300}
k="x"
findKey = False #預設沒有找到key
for key in d3:
    if key == k:
        findKey = True
        break
#找不到key顯示無此key
if findKey:
    print(d3[k])
else:
    print("無此Key")
```

```python
d3 = {"a":100,"b":200,"c":300}
k="z"
for key in d3:
    if key == k:
        print(d3[k])
        break
else:#不是因為break離開迴圈就會執行
    print("無此Key")
```

```python
d3 = {"a":100,"b":200,"c":300}
k = "b"
if k in d3:
    print(d3[k])
else:
    print("無此Key")
```

```python
d3 = {"a":100,"b":200,"c":300}
k = "b"
v = d3[k] if k in d3 else "無此key"
print(v)
```
