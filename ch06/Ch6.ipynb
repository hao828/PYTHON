### 作業：支出最大與最小



```python
money=[]
sum = 0
for i in range(1,5):
    s = input(f"請輸入第{i}個月的支出金額:")
    money.append(s)
    sum += int(s)
money.sort()
print(f"支出最多的金額為:{money[-1]}")
print(f"支出最少的金額為:{money[0]}")
print(f"支出的總金額為:{sum}")
print(f"支出的金額由小到大為:{money}")
```

```python
def printInfo(arg1,arg2="default",*arg3):
    print("arg1:",arg1)
    print("arg2:",arg2)
    print("arg3:",arg3)
    print("==================")
printInfo(1)
printInfo(1,2)
printInfo(1,2,3)
printInfo(1,2,3,4)
```

```python
def printInfo2(*arg3,arg1,arg2="default"):
    print("arg1:",arg1)
    print("arg2:",arg2)
    print("arg3:",arg3)
    print("==================")
#printInfo(arg1=10,arg2=20,3,4)
#printInfo(1,2,3,4,5)
printInfo2(1,2,3,arg1=4,arg2=5)
```

```python
def sumValue(*args):
    v = 0
    for a in args:
        v += a
    return v
print(sumValue())
print(sumValue(2))
print(sumValue(2,3))
print(sumValue(2,3,4))
```

```python
def fun4(arg1,arg2="default",**arg3):
    print(arg1)
    print(arg2)
    print(arg3)
    print("="*20)
fun4(1)
fun4(1,2)
fun4(1,2,name="kevin",age=15,score=86)
```

```python
def calculate(*value,**args):
    cType = "+"
    result = 0
    if "cType" in args:
        cType = args["cType"]
    if cType == "*":
        result = 1
    for v in value:
        if cType == "+":
            result += v
        elif cType == "*":
            result *= v
    return result
r = calculate(3,2,4, cType="*")
print(r)
```

### MAP 適合轉換

```python
list3=[1,2,3,'',4,5,'']
map0b=map(lambda x:x if x!='' else 0,list3)
print(list(map0b))
```

```python
list4 = ['M','W','M','M','W']
myDirc = {'M':0,'W':1}
objMap = map(lambda x:myDirc[x],list4)
print(list(objMap))
```

### Filter 適合過濾

```python
a = [1,2,3,4,5,6]
f1 = filter(lambda x : x%2==0 , a)
print(list(f1))
```

```python
list3 = [1,2,3,'',4,5,'']
f2 = filter(lambda x:x!='',list3)
print(list(f2))
```
