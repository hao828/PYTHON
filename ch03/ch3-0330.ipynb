### 索引標籤與取值


```python
msg = "abcdefghij"
print(len(msg))
print(msg[0])
print(msg[9])
print(msg[5])
print(msg[-3])
print(msg[-1])
```

```python
msg = "abcdefghij"
for i in range(0,len(msg)):
    print(i,msg[i])
```

```python
msg = "abcdefghij"
for i in range(len(msg)-1,-1,-1):
    print(i,msg[i])
```

```python
msg = "abcdefghij"
for i in range(1,4):
    print(msg[i],end="")
```

```python
msg = "abcdefghij"
print(msg[1:4])
```

```python
msg = "abcdefghij"
print(msg[4:7])
print(msg[-3:])
print(msg[:-3])
print(msg[::2]) # 0開始每次index加2
```

```python
msg = "abcdefghij"
msgLan = len(msg)
newMsg = msg[msgLan-1::-1] # 產生新的字串
print(newMsg)
```

```python
msg = "abcdefghij"
newMsg = msg[-3:-8:-2]
print(newMsg)
```

### find


```python
str1 = "This is Python.That is Java,This is SQLite"
print(str1)
search1 = "Python"
print(str1.find(search1))
search2 = "not"
print(str1.find(search2)) # -1表示沒有找到
search3 = "t"
print(str1.find(search3)) # 由左至右開始找，找到後即停止
search4 = "is"
print(str1.find(search4,4)) #可指定從第幾個位置開始找
```

```python
str1 = "This is Python.That is Java,This is SQLite"
search1 = "is"
index = 0
while True:
    index = str1.find(search1,index)
    if(index == -1):
        break
    print("is:",index)
    index += 1
```
